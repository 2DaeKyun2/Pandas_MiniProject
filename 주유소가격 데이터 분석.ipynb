> ### str accessor
> - `Series.str.String메소드()`
> - 문자열 타입 Series에 String 메소드들을 사용하여 일괄처리 할 수 있도록 한다.
> - str accessor 메소드 : https://pandas.pydata.org/pandas-docs/stable/reference/series.html#api-series-str

> ### datetime으로 변환
> - `pd.to_datetime(변환할 Series, format="날짜형식")`
> - x 또는 y 축이 날짜 일 경우 datetime 타입으로 처리하는 것이 좋다.
> - format: 년월일의 형식을 strftime 형식문자열로 지정한다. 
>      - [strftime 형식](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes)  
> - datetime을 변환하면 날짜만의 기능을 다룰 수 있다.
>     - dt accessor를 이용해 datetime의 속성값들을(년, 월, 일 등) 추출할 수 있다.
>     - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.dt.html


```python
# 지역 컬럼의 값을 이용하여 "구" 컬럼 생성 (지역에서 구만 추출해서 파생변수컬럼 생성)

df3['구'] = df3['지역'].apply(lambda x : x.split()[1])
# df3['지역'].str.slice(3) => str accessor 이용

df3.head()

```

```python
# 기간 컬럼을 datetime 타입으로 변경
import datetime

# df3['기간] = pd.to_datetime(df3['기간'], format="%Y%m%d")      # Y:년, m:월 d:일, H: 시간, M: 분, s: 초
df3['기간'] = df3['기간'].astype('str')


df3['기간'] = pd.to_datetime(df3['기간'])

df3.info()
```

```python
# 기간 컬럼을 이용해 "월", "일", "요일" 컬럼 생성 (dt accessor 이용)

df3['월'] = df3['기간'].dt.month
df3['일'] = df3['기간'].dt.day
df3['요일'] = df3['기간'].dt.weekday      # dt.month(월), dt.day(일)
df3['요일'] = df3['요일'].apply(lambda x : '월화수목금토일'[x])
```

```python
# 가장 휘발유가격이 비싼 주유소 상위 5 - (같은 주유소가 여러개 나오므로 가장 비싼 가격을 기준으로 집계)
# df3.loc[df3['휘발유'].sort_values(ascending=False).head(5).index]

# 상호, 주소 나눔 => 상호가 같은 주유소가 있을 경우 대비
expensive_oil = df3.groupby(['상호', '주소'])['휘발유'].max().sort_values(ascending=False).head(5)
expensive_oil
```

```python
# 가장 휘발유가격이 저렴한 주유소 상위 5 - (같은 주유소가 여러개 나오므로 가장 '비싼 가격'을 기준으로 집계)
cheep_oil = df3.query('휘발유!=0').groupby(['상호', '주소'])['휘발유'].max().sort_values().head(5)
cheep_oil
```

```python
df3.query('상호 == "송파알찬주유소"')[['상호', '주소', '지역', '상표']].drop_duplicates()       # 중복 행 제거
```

```python
# index나 컬럼명이 여러개 => Multi-Index
# 특정 level의 index 조회 (level -> 밖에서 안쪽으로 0에서 1씩 증가. (음수 index도 있다.))
cheep_oil.index.get_level_values(0)     # level 0 인 index값 조회
cheep_oil.index.get_level_values(-1)    # 맨 안쪽 index
```

```python
# 휘발유 비싼 주유소의 상위 5개의 상호, 지역, 주소 조회(중복제거-drop_duplicates) ==> 비싼 주유소 조회결과에 추가정보 확인

df3.loc[df3['상호'].isin(expensive_oil.index.get_level_values(0)) & 
        df3['주소'].isin(expensive_oil.index.get_level_values(1)), ['상호', '지역', '주소']].drop_duplicates()

```

```python
# 휘발유 저렴한 주유소의 상위 5개의 상호, 지역, 주소 조회(중복제거)
df3.query('상호 in @cheep_oil.index.get_level_values(0) and 주소 in @cheep_oil.index.get_level_values(1)')[['상호', '지역', '주소']].drop_duplicates()
```
